## ciclos while

```python
# Acumulador de valores con límite definido
# Este script utiliza un contador para solicitar una cantidad específica de números y sumarlos en una variable totalizadora

# El usuario define cuántas vueltas dará el ciclo
vueltas = int(input("¿Cuántos números sumarás? "))
suma_total = 0
contador = 0
# El ciclo se controla estrictamente por la comparación del contador
while contador < vueltas:
    valor = int(input("Ingresa un número: "))
    suma_total += valor
    contador += 1
```

```python
# Recolección de datos en una lista
# Aquí se utiliza el ciclo para pedir elementos y guardarlos en una estructura de datos (lista) hasta alcanzar el tamaño deseado por el usuario

# Se solicita el tamaño de la colección
limite = int(input("¿Cuántos nombres guardarás? "))
lista_nombres = []
# La condición del while determina el final sin necesidad de un if
while len(lista_nombres) < limite:
    nombre = input("Ingresa un nombre: ")
    lista_nombres.append(nombre)
```

```python
# Validación de coincidencia exacta (Centinela)
# Este código se repite indefinidamente mientras la entrada del usuario sea distinta a una clave o palabra secreta definida

clave_maestra = "Python2026"
intento = input("Ingresa la clave de acceso: ")
# La lógica de "distinto de" (!=) mantiene el bucle activo
while intento != clave_maestra:
    intento = input("Clave incorrecta. Intenta de nuevo: ")
```

```python
# Búsqueda de potencia matemática
# El ciclo incrementa un valor multiplicándolo por sí mismo hasta que supera un umbral ingresado por el usuario

limite_numerico = int(input("Ingresa el tope para la potencia de 2: "))
resultado_potencia = 1
# El bucle busca el primer valor que rompa la condición de ser menor al límite
while resultado_potencia < limite_numerico:
    resultado_potencia *= 2
```

```python
# Generador de cadena por longitud
# Este ejemplo solicita un carácter y lo concatena consigo mismo en una variable hasta que la longitud de la cadena alcanza la meta establecida

caracter = input("Ingresa un carácter para repetir: ")
meta_longitud = int(input("¿Qué longitud debe tener la cadena final? "))
cadena_resultante = caracter
# El ciclo se detiene automáticamente cuando la longitud iguala o supera la meta
while len(cadena_resultante) < meta_longitud:
    cadena_resultante += caracter
```
